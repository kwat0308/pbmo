```python
from pbmo.pbmo import PBMO
import numpy as np
```

```python
# initialize row / columns
rowdims = np.arange(5, 15, 5)
coldims = np.arange(5, 15, 5)
dims = list(zip(rowdims, coldims))
```

```python
# initialize benchmark solver
pbmo = PBMO(dims=dims)
```

```python
# evaluate norm
pbmo.evaluate_norm(show_progress=True)
```

```python
pbmo.collect_results()
```

```python
pbmo.print_results()
```

```python
print(list(pbmo.results["Python"].values())[0][1])
```

```python
pbmo.plot_results(plot_ratios=True)
```

```python
print([pbmo.results[mat_type]["{0} Time".format(
                "Norm")] for mat_type in pbmo.matrix_types])
```
